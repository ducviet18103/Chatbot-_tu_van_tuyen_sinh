```python
import psutil
import os

def find_processes_using_file(filepath):
    for proc in psutil.process_iter(['pid', 'name', 'open_files']):
        try:
            if proc.open_files():
                for file in proc.open_files():
                    if filepath in file.path:
                        print(f"Process using the file: {proc.name()} (PID: {proc.pid})")
        except (psutil.NoSuchProcess, psutil.AccessDenied):
            pass

# Check what's using the lock file
find_processes_using_file('domainknowledge\\.lock')
```
