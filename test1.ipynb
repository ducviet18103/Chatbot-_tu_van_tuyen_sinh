```python
import os
from docx import Document

def convert_docx_to_txt(input_file, output_file):
    # Load the .docx file
    document = Document(input_file)
    
    # Extract text from each paragraph
    text = '\n'.join([paragraph.text for paragraph in document.paragraphs])
    
    # Save the extracted text to a .txt file
    with open(output_file, 'w', encoding='utf-8') as txt_file:
        txt_file.write(text)
    print(f"Conversion completed. Text saved to {output_file}")

# Convert all .docx files in the raw_data folder to .txt
raw_data_folder = 'raw_data'
for filename in os.listdir(raw_data_folder):
    if filename.endswith('.docx'):
        input_path = os.path.join(raw_data_folder, filename)
        # output_path = os.path.join(raw_data_folder, filename.replace('.docx', '.txt'))
        # convert_docx_to_txt(input_path, output_path)
        os.remove(input_path)  # Delete the .docx file after conversion
        print(f"Deleted {input_path}")
```
